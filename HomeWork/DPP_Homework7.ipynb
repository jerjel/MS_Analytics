Suppose you have a dataframe <i>dfdi</i> with the features of diamonds.

```python
from seaborn import load_dataset
dfdi = load_dataset("diamonds")
dfdi.head()
```

For questions 1 to 6, let's aim to build a <b>regression</b> model using the diamond dataset that is able to predict the price of a new diamond based on its carat, cut, color, clarity, depth, table, x, y, and z. 

1\. There are 3 categorical columns in <i>dfdi</i>: <i>cut</i>, <i>color</i>, and <i>clarity</i>. Convert those columns into numerical columns with the corresponding __category codes__. (10 pts)

```python
# Your answer here
import pandas as pd    
dfdi.cut = pd.Categorical(dfdi.cut) 
dfdi.color = pd.Categorical(dfdi.color) 
dfdi.clarity = pd.Categorical(dfdi.clarity) 
dfdi.cut = dfdi.cut.cat.codes 
dfdi.color = dfdi.color.cat.codes 
dfdi.clarity = dfdi.clarity.cat.codes 
dfdi.head()
```

2\. According to the goal description above, there are 9 features and 1 target. Based on those features and target, define the two variables <i>X</i> as a feature dataset and <i>y</i> as a target dataset. (10 pts)

```python
# Your answer here
features = list(dfdi.columns)
features.remove('price')
features
target = "price"
X = dfdi[features]
y = dfdi[target]
```

3\. Split <i>X</i> and <i>y</i> into two training datasets <i>X_train</i> and <i>y_train</i> and two test datasets <i>X_text</i> and <i>y_test</i>. Set the `test_size` to 0.2 and `random_state` to 0. (10 pts)

```python
# Your answer here
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

4\. Build a new linear regression model and fit the model using the two training datasets. (10 pts)

```python
# Your answer here
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train, y_train)
```

5\. Get the training score and test score, respectively. (10 pts)

```python
# Your answer here
print('Training Score ',lr.score(X_train, y_train))
print('Testing Score ',lr.score(X_test, y_test))
```

6\. Suppose there is a new diamond that has the following features. How much would the price be using your linear regression model? (10 pts)

```python
d = {"carat": 0.7,
     "cut": 4,
     "color": 1,
     "clarity": 5,
     "depth": 62.4,
     "table": 60.0,
     "x": 5.57,
     "y": 5.61,
     "z": 3.49}

new_diamond = [d["carat"], d["cut"], d["color"], d["clarity"], d["depth"], d["table"], d["x"], d["y"], d["z"]]
X_new = [new_diamond]
```

```python
# Your answer here
lr.predict(X_new)
```

__For questions 7 to 10__, using the same diamond data, let's aim to build a <b>clustering</b> model that is able to group all diamonds into 10 clusters of similar ones by looking at the same 9 features above along with the target. 

We need a copy of <i>dfdi</i> for clustering. Use <i>dfdic</i> for your clustering.

```python
dfdic = dfdi.copy()
```

For clustering, all you need is just an input dataset <i>X</i>. Let's use all columns in <i>dfdic</i> as features.

```python
X = dfdic
```

7\. Build a new k-means clustering model <i>kmeans</i>. Set `n_clusters` according to the goal description and `random_state` to 0. Fit the clustering model using the input dataset <i>X</i>. (10 pts)

```python
# Your answer here
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=0)  
kmeans.fit(X)
```

8\. Assign the resulting labels of <i>kmeans</i> to the new column of <i>dfdic</i> called <i>label</i>. (10 pts)

```python
# Your answer here
dfdic["label"] = kmeans.labels_
```

9\. Check the number of values for each label. (10 pts)

```python
# Your answer here
dfdic.label.value_counts()  
```

10\. Select a random sample of <i>dfdic</i> with 30 rows that have the lable 4. For random sampling, set `replace` to False and `random_state` to 0. (10 pts)

```python
# Your answer here
dfdic[dfdic.label == 4].sample(n=30, replace=False, random_state=0) 
```

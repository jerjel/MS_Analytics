9\. Get the mean price for each color and store in a dictionary ans9, which has colors as the key and mean results as the value. (10 pts)

```python
# Your answer here
ans9={}
dff1 = df.groupby(["color"]).price.mean()

for key, value in dff1.iteritems():
    ans9[key]=value
```

4\.
```python
# Your answer here
import json
diamonds=[]
df_new.columns = df_new.columns.str.replace(' ', '')
# df_new.carat = df_new.carat.astype(float)
# df_new.cut = df_new.cut.astype(str)
# df_new.color = df_new.color.astype(str)
# df_new.clarity = df_new.clarity.astype(str)
# df_new.depth = df_new.depth.astype(float)
# df_new.table = df_new.table.astype(int)

# df_new['table'] = df_new['table'].astype('int')
# df_new.price = df_new.price.astype(int)

for i in range(len(df_new)):
    diamond={}
    for c in df_new.columns:
        diamond[c] = df_new[c][i]
    diamonds.append(diamond)
    
for d in diamonds:
    d["carat"] = str(d["carat"])
    d["depth"] = str(d["depth"])
    d["table"] = str(d["table"])
    d["price"] = str(d["price"])
    
with open("hw_diamonds.json", "w") as fw:
    json.dump(diamonds, fw)
```

7\. Get the mean PressureRange for each WindGustDir in a dictionary ans7, which has WindGustDir as the key and mean results as the value. You will need to update any entries missing a WindGustDirection with the most frequent value. (15 pts)

```python
# Your answer here
ans7 = {}
weather_df["WindGustDir"] = weather_df["WindGustDir"].fillna(weather_df["WindGustDir"].value_counts().idxmax())
df_pressure_range = weather_df.groupby(["WindGustDir"]).PressureRange.mean()
for key, value in df_pressure_range.iteritems():
    ans7[key]=value
```

8\. Add a <i>Month</i> column to weather_df that is the abbreviated text representation of Month (e.g. 'Jan', 'Feb', 'Mar') as a __categorical__ variable. Also, add a boolean column called <i>BusDay</i> that indicates whether the day is a business day or not. (10 pts)
